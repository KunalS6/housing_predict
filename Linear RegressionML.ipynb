```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def fillblank(table_name,column_name):
    mean_val=table_name[column_name].mean()
    return table_name[column_name].fillna(value=mean_val,inplace=True)

def relation_variable(table_name,col1_name,col1_label,col2_name,col_label):
    plt.scatter(table_name[col1_name],table_name[col2_name])
    plt.xlabel(col1_label)
    plt.ylabel(col_label)
    return plt.show()
 
def regression_plot(table_name,col1,col2):
    return sns.regplot(x=col1,y=col2,data=table_name)
```

```python
fillblank(dataset,'CRIM')
fillblank(dataset,'ZN')
fillblank(dataset,'INDUS')
fillblank(dataset,'CHAS')
fillblank(dataset,'AGE')
fillblank(dataset,'LSTAT')
```

```python
X=dataset.iloc[:,:-1]
```

```python
y=dataset.iloc[:,-1]
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
#which param used in the model
reg.get_params()
```

```python
#predict the data
reg_pred=reg.predict(X_test)
```

```python
plt.scatter(y_test,reg_pred)
```

```python
#residual 
sns.displot(y_test-reg_pred,kind='kde')
```

```python
residual=y_test-reg_pred
```

```python
plt.scatter(reg_pred,residual)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
print('MAE: ',mean_absolute_error(y_test,reg_pred))
print('MSE: ',mean_squared_error(y_test,reg_pred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
#more toward the 1 more better for the model
```

```python
adjusted_r2=1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
adjusted_r2
#that needs to be less than r2
```
